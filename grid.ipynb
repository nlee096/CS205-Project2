```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('iris.txt', delimiter=' ')
features = data[:,1:]
label = data[:,0]
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(10,10))
labels = ["Iris Setosa", "Iris Versicolour", "Iris Virginica"]
for i, ax_row in enumerate(axes, start=1): 
    x1=np.array([])
    x2=np.array([])
    x3=np.array([])
    for instance_x in range(150):
        if label[instance_x] == 1:
            x1 = np.append(x1, features[instance_x,i-1])
        elif label[instance_x] == 2:
            x2 = np.append(x2, features[instance_x,i-1])
        elif label[instance_x] == 3:
            x3 = np.append(x3, features[instance_x,i-1])
        else:
            print("instance has unrecognized class", instance_x, label[instance_x])
    if i == 1:
        xlabel = "Sepal Length (cm)"
    elif i == 2:
        xlabel = "Sepal Width (cm)"
    elif i == 3:
        xlabel = "Petal Length (cm)"
    elif i == 4:
        xlabel = "Petal Width (cm)"
    else:
        print("feature number unrecognized")
    for j, ax in enumerate(ax_row, start=1):
        ax.set_title("Feature " + str(j) + " vs Feature " + str(i))
        y1=np.array([])
        y2=np.array([])
        y3=np.array([])
        for instance_y in range(150):
            if label[instance_y] == 1:
                y1 = np.append(y1, features[instance_y,j-1])
            elif label[instance_y] == 2:
                y2 = np.append(y2, features[instance_y,j-1])
            elif label[instance_y] == 3:
                y3 = np.append(y3, features[instance_y,j-1])
            else:
                print("instance has unrecognized class", instance_y, label[instance_y])
        if j == 1:
            ylabel = "Sepal Length (cm)"
        elif j == 2:
            ylabel = "Sepal Width (cm)"
        elif j == 3:
            ylabel = "Petal Length (cm)"
        elif j == 4:
            ylabel = "Petal Width (cm)"
        else:
            print("feature number unrecognized")
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        ax.scatter(x1,y1, c="red")
        ax.scatter(x2,y2, c="blue")
        ax.scatter(x3,y3, c="green")

fig.legend(labels=labels, loc="upper right", bbox_to_anchor=(1.18, 1))
fig.tight_layout(h_pad=3)
plt.show()
```
